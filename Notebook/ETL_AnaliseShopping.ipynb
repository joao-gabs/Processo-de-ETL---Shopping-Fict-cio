```python
import mysql.connector
import pandas as pd
import json

with open('C:/Users/Felippe/OneDrive/Desktop/Projetos/Projeto ETL/config.json','r') as config_file:
    config = json.load(config_file)

connection = mysql.connector.connect(
    host = config['db_host'],
    user = config['db_user'],
    password = config['db_password'],
    database = config['db_name']
)

consulta_tabela = "SELECT * FROM customer_purchases"
df = pd.read_sql(consulta_tabela,connection)

connection.close()

display(df.head())
```

```python
# Valores nulos no dataset.

display(df.isnull().sum())

# Valores duplicados

duplicated_rows = df[df.duplicated()]
display(duplicated_rows)


```

### Este dataset está quase tratado, não tem duplicatas e nem valores nulos, assim como os tipos de dados são convenientes em cada categoria no conjunto!
### Porém vamos remover algumas colunas e renomear para o arquivo final!



```python
#Remoção das colunas que queremos aplicar

remove_columns = [0,4,6,7,8,9,10,11,13,14,15,17]
df.drop(df.columns[remove_columns], axis=1, inplace=True)


df.rename(columns={"age": "Idade","gender":"Genero","item_purchased": "Item_Comprado", "purchase_amount": 'Valor_Compra(R$)', "payment_method": "Forma_Pagamento","previous_purchases": "Compras_Anteriores", "frequency_of_purchases":"Frequencia_Compras"},inplace=True)
df['Genero'] = df['Genero'].map({'Male': 'Masculino', 'Female': 'Feminino'})
df['Item_Comprado'] = df['Item_Comprado'].map({'Blouse': 'Blusa', 'Sweater': 'Sueter','Jeans': 'Jeans', 'Sandals': 'Sandalias', 'Sneakers': 'Tenis', 'Shirt': 'Camisa', 'Shorts':'Shorts','Coat': 'Casaco',
                                              'Handbag': 'Bolsa','Shoes':'Sapato','Dress': 'Vestido', 'Skirt': 'Saia','Sunglasses': 'Óculos de Sol', 'Pants': 'Calca', 'Jacket': 'Jaqueta','Hoodie': 'Agasalho',
                                              'Jewelry': 'Joias','T-shirt': 'Camiseta','Scarf':'Cachecol','Hat': 'Bone','Socks': 'Meias', 'Backpack':'Mochila','Belt':'Cinto','Boots': 'Botas','Gloves':'Luvas'})

df['Forma_Pagamento'] = df['Forma_Pagamento'].map({'Credit Card':'Cartao de Credito','Bank Transfer':'Transferencia Bancaria','Cash':'Dinheiro','PayPal':'PayPal','Venmo':'Venmo','Debit Card': 'Cartao de Debito'})
df['Frequencia_Compras'] = df['Frequencia_Compras'].map({'Fortnightly':'Quinzenal','Weekly':'Semanal','Annually':'Anual', 'Quarterly':'Trimestral','Bi-Weekly':'Duas vezes/semana','Monthly':'Mensal','Every 3 Months':'A cada 3 meses'})

display(df)
```

## Com os dados devidamente tratados e padronizados, podemos converter este arquivo para csv novamente e fazer a manipulação para a visualização de dados com o Power BI.

```python
df.insert(0, 'Id', range(1, len(df) + 1))


df.to_csv('AnaliseShopping.csv', index=False, encoding='utf-8')
print("Arquivo CSV salvo com sucesso!")
```
